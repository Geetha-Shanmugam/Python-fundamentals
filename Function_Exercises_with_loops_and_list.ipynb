To check the average error

```python
def AI_testing_average_error(error_list):
    total_error = 0
    for a in error_list:
     total_error = a + total_error
    return total_error/len(error_list)
my_data = [2, 5, 8]
total_error = AI_testing_average_error(my_data)
print(total_error)
```
